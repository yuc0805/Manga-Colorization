```python
import cv2
import os

# Define the input and output directories
input_dir = '/Users/leo/Desktop/Manga-Colorization/Manga_CycleGAN/manga/panels/'
output_dir = '/Users/leo/Desktop/Manga-Colorization/Manga_CycleGAN/manga/gray_panels/'

# Create the output directory if it does not exist
os.makedirs(output_dir, exist_ok=True)

# Loop through all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.png') or filename.endswith('.jpg') or filename.endswith('.jpeg'):
        # Load the image
        image_path = os.path.join(input_dir, filename)
        image = cv2.imread(image_path)
        
        # Check if the image is loaded successfully
        if image is not None:
            print('Processing image:', filename, 'with shape:', image.shape)
            
            # Convert the image to grayscale
            image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            
            # Save the grayscale image to the output directory
            output_path = os.path.join(output_dir, filename)
            cv2.imwrite(output_path, image_gray)
        else:
            print('Failed to load image:', filename)

print('Conversion completed.')
```
